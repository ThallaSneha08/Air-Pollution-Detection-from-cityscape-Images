```python
data_dir = "data/Dataset_for_AQI_Classification"
print("‚úÖ Dataset ready at:", data_dir)
!ls {data_dir}
```

```python
IMG_SIZE = (224, 224)
BATCH = 32
EPOCHS = 15

train_gen = ImageDataGenerator(
    rescale=1./255,
    validation_split=0.2,
    rotation_range=10,
    width_shift_range=0.1,
    height_shift_range=0.1,
    zoom_range=0.1,
    horizontal_flip=True,
    brightness_range=(0.7,1.3),
    fill_mode='nearest'
)
```

```python
train_ds = train_gen.flow_from_directory(
    data_dir, target_size=IMG_SIZE, batch_size=BATCH, subset='training', class_mode='categorical'
)
```

```python
val_ds = train_gen.flow_from_directory(
    data_dir, target_size=IMG_SIZE, batch_size=BATCH, subset='validation', class_mode='categorical'
)
```

```python
NUM_CLASSES = len(train_ds.class_indices)
print("Classes:", train_ds.class_indices)
```

```python
base = EfficientNetB0(include_top=False, input_shape=(*IMG_SIZE,3), weights='imagenet')
base.trainable = False
```

```python
inputs = layers.Input(shape=(*IMG_SIZE,3))
x = base(inputs, training=False)
x = layers.GlobalAveragePooling2D()(x)
x = layers.Dropout(0.4)(x)
x = layers.Dense(128, activation='relu')(x)
x = layers.Dropout(0.3)(x)
outputs = layers.Dense(NUM_CLASSES, activation='softmax')(x)
model = models.Model(inputs, outputs)
```

```python
model.compile(optimizer=tf.keras.optimizers.Adam(1e-3),
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```
